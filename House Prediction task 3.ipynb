```python
import numpy as np
import pandas as pd
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = fetch_california_housing(as_frame=True)
```

```python
df = pd.concat([data.data, data.target.rename('MedHouseVal')], axis=1)
df.head()
```

```python
Y = df['MedHouseVal']
X = df.drop(labels="MedHouseVal",axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()      
X_scaled = scaler.fit_transform(X)
```

```python
x_train, x_test, Y_train, Y_test = train_test_split(
    X_scaled,
    Y,
    test_size=0.2,
    random_state=42
)

print(x_train.shape)
print(x_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Ridge Regression": Ridge(alpha=1.0)
}
```

```python
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error, r2_score


results = {}


for name, model in models.items():

    
    model.fit(x_train, Y_train)

    
    y_pred = model.predict(x_test)

   
    rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
    r2 = r2_score(Y_test, y_pred)

   
    results[name] = {
        "RMSE": rmse,
        "R2 Score": r2
    }


results_df = pd.DataFrame(results).T


results_df = results_df.sort_values(by="R2 Score", ascending=False)

print(results_df)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error

# Create the model (high chance of overfitting)
model = DecisionTreeRegressor(max_depth=None, random_state=42)

# Train the model
model.fit(x_train, Y_train)

# Predictions
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)

# R2 Scores
train_r2 = r2_score(Y_train, train_pred)
test_r2 = r2_score(Y_test, test_pred)

# RMSE
train_rmse = np.sqrt(mean_squared_error(Y_train, train_pred))
test_rmse = np.sqrt(mean_squared_error(Y_test, test_pred))

print("Decision Tree Regressor")
print("Train R2 Score :", train_r2)
print("Test R2 Score  :", test_r2)
print("Train RMSE     :", train_rmse)
print("Test RMSE      :", test_rmse)

# Check Overfitting
if train_r2 > test_r2 + 0.10:
    print("\nModel is OVERFITTING")
else:
    print("\nModel is not significantly overfitting")
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error

# Create the model (high chance of overfitting)
model = Ridge(alpha=1.0)

# Train the model
model.fit(x_train, Y_train)

# Predictions
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)

# R2 Scores
train_r2 = r2_score(Y_train, train_pred)
test_r2 = r2_score(Y_test, test_pred)

# RMSE
train_rmse = np.sqrt(mean_squared_error(Y_train, train_pred))
test_rmse = np.sqrt(mean_squared_error(Y_test, test_pred))

print("Ridge Regressor")
print("Train R2 Score :", train_r2)
print("Test R2 Score  :", test_r2)
print("Train RMSE     :", train_rmse)
print("Test RMSE      :", test_rmse)

# Check Overfitting
if train_r2 > test_r2 + 0.10:
    print("\nModel is OVERFITTING")
else:
    print("\nModel is not significantly overfitting")
```

```python
## LIMIT TREE DEPTH
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# Controlled Decision Tree
controlled_model = DecisionTreeRegressor(
    max_depth=5,
    random_state=42
)

controlled_model.fit(x_train, Y_train)

train_pred = controlled_model.predict(x_train)
test_pred = controlled_model.predict(x_test)

train_r2 = r2_score(Y_train, train_pred)
test_r2 = r2_score(Y_test, test_pred)

print("After Controlling Overfitting")
print("Train R2:", train_r2)
print("Test R2 :", test_r2)
```

```python
#Comparing before and after the limit depth tree
import pandas as pd

comparison = pd.DataFrame({
    "Model": ["Original Decision Tree", "Controlled Decision Tree"],
    "Train R2": [1.00, train_r2],
    "Test R2": [0.62, test_r2]
})

print(comparison)
```

```python
## Cross Validation
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    controlled_model,
    X,
    Y,
    cv=5,
    scoring='r2'
)

print("Cross Validation Scores:", cv_scores)
print("Average CV R2:", cv_scores.mean())
```

```python
#Hyper parameter tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10]
}

grid = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring='r2'
)

grid.fit(x_train, Y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_

y_pred = best_model.predict(x_test)

r2 = r2_score(Y_test, y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, y_pred))

print("rmse  :", rmse)
print("r2  :", r2)
```
